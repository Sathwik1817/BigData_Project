```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.DataFrame(pd.read_csv("/content/drive/MyDrive/Colab Notebooks/csh101 ann.csv"))
```

```python
df
```

```python
column=['Sensor', 'Translate01','Translate02', 'Message', 'SensorType','Unnamed: 8','Unnamed: 9','Unnamed: 10']
df=df.drop(column, axis=1)
new=pd.to_datetime(df.Date.astype(str)+ ' '+df.Time.astype(str))
df.insert(0,"Timestamp",new)
df
```

```python
Activity_counts=df['Activity'].value_counts()
Activity_counts
```

```python
Watch_TV_df=df.query("Activity=='Watch_TV'")
array=np.array(Watch_TV_df['Time'])
array2=[]
for i in array:
  a=i.split(":")
  i=a[0]+":00:00"
  i=a[0]+":00:00"
  array2.append(i)
Watch_TV_df= Watch_TV_df.assign(Time=array2)
Watch_TV_df["Time"] = pd.to_datetime(Watch_TV_df["Time"], format='%H:%M:%S').apply(lambda x: x.time())
Watch_TV_df
```

```python
Bathe_df=df.query("Activity=='Bathe'")
array=np.array(Bathe_df['Time'])
array2=[]
for i in array:
  a=i.split(":")
  i=a[0]+":00:00"
  i=a[0]+":00:00"
  array2.append(i)
Bathe_df= Bathe_df.assign(Time=array2)
Bathe_df["Time"] = pd.to_datetime(Bathe_df["Time"], format='%H:%M:%S').apply(lambda x: x.time())
Bathe_df
```

```python
Personal_Hygiene_df=df.query("Activity=='Personal_Hygiene'")
array=np.array(Personal_Hygiene_df['Time'])
array2=[]
for i in array:
  a=i.split(":")
  i=a[0]+":00:00"
  i=a[0]+":00:00"
  array2.append(i)
Personal_Hygiene_df= Personal_Hygiene_df.assign(Time=array2)
Personal_Hygiene_df["Time"] = pd.to_datetime(Personal_Hygiene_df["Time"], format='%H:%M:%S').apply(lambda x: x.time())
Personal_Hygiene_df
```

```python
Sleep_df=df.query("Activity=='Sleep'")
array=np.array(Sleep_df['Time'])
array2=[]
for i in array:
  a=i.split(":")
  i=a[0]+":00:00"
  i=a[0]+":00:00"
  array2.append(i)
Sleep_df= Sleep_df.assign(Time=array2)
Sleep_df["Time"] = pd.to_datetime(Sleep_df["Time"], format='%H:%M:%S').apply(lambda x: x.time())
Sleep_df
```

```python
Phone_df=df.query("Activity=='Phone'")
array=np.array(Phone_df['Time'])
array2=[]
for i in array:
  a=i.split(":")
  i=a[0]+":00:00"
  i=a[0]+":00:00"
  array2.append(i)
Phone_df= Phone_df.assign(Time=array2)
Phone_df["Time"] = pd.to_datetime(Phone_df["Time"], format='%H:%M:%S').apply(lambda x: x.time())
Phone_df
```

```python
Watch_TV_pivot = pd.pivot_table(Watch_TV_df, values='Activity', index="Date", columns="Time",aggfunc='count',fill_value=0)
Bathe_pivot = pd.pivot_table(Bathe_df, values='Activity', index="Date", columns="Time",aggfunc='count',fill_value=0)
Personal_Hygiene_pivot = pd.pivot_table(Personal_Hygiene_df, values='Activity', index="Date", columns="Time",aggfunc='count',fill_value=0)
Sleep_pivot = pd.pivot_table(Sleep_df, values='Activity', index="Date", columns="Time",aggfunc='count',fill_value=0)
Phone_pivot = pd.pivot_table(Phone_df, values='Activity', index="Date", columns="Time",aggfunc='count',fill_value=0)
```

```python
Watch_TV_Time = Watch_TV_pivot.iloc[0].index
Bathe_Time = Bathe_pivot.iloc[0].index
Personal_Hygiene_Time = Personal_Hygiene_pivot.iloc[0].index
Sleep_Time = Sleep_pivot.iloc[0].index
Phone_Time = Phone_pivot.iloc[0].index
```

```python
Watch_tv_df = Watch_TV_pivot.reset_index()
Watch_tv_df.head()
```

```python
Bathe_df = Bathe_pivot.reset_index()
Bathe_df.head()
```

```python
Personal_Hygiene_df = Personal_Hygiene_pivot.reset_index()
Personal_Hygiene_df.head()
```

```python
Sleep_df = Sleep_pivot.reset_index()
Sleep_df.head()
```

```python
Phone_df = Phone_pivot.reset_index()
Phone_df.head()
```

```python
Watch_TV_hours=[]
x=0
for col in Watch_TV_Time:
  if col !='Date':
    x = Watch_TV_pivot[col].tolist()
  Watch_TV_hours.append(x)
Bathe_hours = []
x = 0
for col in Bathe_Time:
  if col != 'Date':
    x = Bathe_pivot[col].tolist()
  Bathe_hours.append(x)
Personal_Hygiene_hours = []
x = 0
for col in Personal_Hygiene_Time:
  if col != 'Date':
    x = Personal_Hygiene_pivot[col].tolist()
  Personal_Hygiene_hours.append(x)
Sleep_hours = []
x = 0
for col in Sleep_Time:
  if col != 'Date':
    x = Sleep_pivot[col].tolist()
  Sleep_hours.append(x)
Phone_hours = []
x = 0
for col in Phone_Time:
  if col != 'Date':
    x = Phone_pivot[col].tolist()
  Phone_hours.append(x)
```

```python
Watch_TV_Dates=Watch_tv_df['Date'].tolist()
Bathe_Dates = Bathe_df['Date'].tolist()
Personal_Hygiene_Dates = Personal_Hygiene_df['Date'].tolist()
Sleep_Dates = Sleep_df['Date'].tolist()
Phone_Dates = Phone_df['Date'].tolist()
```

```python
Watch_tv_X=[]
Watch_tv_Y=[]
Bathe_X = []
Bathe_Y = []
Personal_Hygiene_X = []
Personal_Hygiene_Y = []
Sleep_X = []
Sleep_Y = []
Phone_X = []
Phone_Y = []
```

```python
for x in Watch_TV_hours:
  t=max(x)
  index=x.index(t)
  Watch_tv_X.append(t)
  Watch_tv_Y.append(Watch_TV_Dates[index])
for x in Bathe_hours:
  t = max(x)
  index = x.index(t)
  Bathe_X.append(t)
  Bathe_Y.append(Bathe_Dates[index])
for x in Personal_Hygiene_hours:
  t = max(x)
  index = x.index(t)
  Personal_Hygiene_X.append(t)
  Personal_Hygiene_Y.append(Personal_Hygiene_Dates[index])
for x in Phone_hours:
  t = max(x)
  index = x.index(t)
  Phone_X.append(t)
  Phone_Y.append(Phone_Dates[index])
for x in Sleep_hours:
  t = max(x)
  index = x.index(t)
  Sleep_X.append(t)
  Sleep_Y.append(Sleep_Dates[index]) 
```

```python
Watch_TV = pd.DataFrame({'Time': Watch_TV_Time, 'Count': Watch_tv_X, 'Date': Watch_tv_Y})
Bathe = pd.DataFrame({'Time': Bathe_Time, 'Count': Bathe_X, 'Date': Bathe_Y})
Personal_Hygiene = pd.DataFrame({'Time': Personal_Hygiene_Time, 'Count': Personal_Hygiene_X, 'Date': Personal_Hygiene_Y})
Sleep = pd.DataFrame({'Time': Sleep_Time, 'Count': Sleep_X, 'Date': Sleep_Y})
Phone = pd.DataFrame({'Time': Phone_Time, 'Count': Phone_X, 'Date': Phone_Y})
```
